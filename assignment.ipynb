```python
# create a method that takes the following variables:
#height = 1.8
#Weight = 55
#now create the BMI of the patient 
# then print out wether the patient has a high  low  or normalBMI based on the outcome.  ( High is 30, Low is under 10)

def calc_bmi():
    #Define the variables
    height = 1.8   # meters
    weight = 55    # kilograms
    
    #Calculate BMI (weight divided by height squared)
    bmi = weight / (height ** 2)
    
    #Print the BMI value
    print(f"Calculated BMI: {bmi:.2f}")
    
    #Determine BMI category based on given thresholds
    if bmi >= 30:
        print("The patient has a High BMI")
    elif bmi < 10:
        print("The patient has a Low BMI")
    else:
        print("The patient has a Normal BMI")
    
    #Return the BMI value
    return bmi
	
#Call Function
calc_bmi()
```
